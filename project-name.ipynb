```python
# Import data, convert string dates to 'datetime64' and set the date column as index:

df = pd.read_csv('/kaggle/input/metro-train-dataset/MetroPT3(AirCompressor).csv',
                 parse_dates=['timestamp'],
                 infer_datetime_format=True,
                 index_col='timestamp',
                 thousands=',',
                 decimal='.'
                )
```

# " Visualization into different Plots "

```python
# Scatter plot

plt.scatter(df['Oil_level'], df['Oil_temperature'])
plt.xlabel('Oil_level')
plt.ylabel('Oil_temperature')
plt.title('Oil Level vs Oil Temperature')
plt.show()
```

```python
# Plot the time series
plt.style.use('fivethirtyeight')
df.plot(subplots=True,
        layout=(6, 3),
        figsize=(22,22),
        fontsize=10, 
        linewidth=1,
        sharex=False,
        title='Visualization of the original Time Series')
plt.show()

# the sharex parameter is used to control the sharing of the x-axis between subplots. 
# It determines whether the subplots in a grid share the same x-axis scale or 
# have independent x-axis scales.
```

```python
# Plot the time series

plt.style.use('fivethirtyeight')
df.plot(kind='kde',
        subplots=True,
        layout=(6, 3),
        figsize=(22,22),
        fontsize=10, 
        linewidth=2,
        sharex=False,
        title='Visualization of the original Time Series')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['TP2'], df['H1'])
plt.xlabel('TP2')
plt.ylabel('H1')
plt.title('TP3 vs H1')
plt.show()
```

```python
# Sin Wave Plot
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin Wave Plot')
plt.show()
```

```python
# Let's also draw a heatmap visualization of the correlation matrix

corr_matrix = df.corr(method='spearman')
f, ax = plt.subplots(figsize=(18,9))
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f', 
            linewidth=0.,
            annot_kws={"size": 10}, 
            cmap='coolwarm', ax=ax)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Histogram Plot
plt.figure(figsize=(8, 6))
plt.hist(df['TP2'], bins=10)
plt.xlabel('TP2')
plt.ylabel('Frequency')
plt.title('Histogram Plot: TP2')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(df[['TP2', 'TP3', 'H1']].values, labels=['TP2', 'TP3', 'H1'])
plt.title('Box Plot: TP2, TP3, H1')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6, 6))
pie_data = df['COMP'].value_counts()
plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%')
plt.title('Pie Chart: COMP')
plt.show()
```

```python
# Joint Plot
plt.figure(figsize=(18, 24))
sns.jointplot(x='Motor_current', y='Oil_temperature', data=df, kind='scatter')
plt.title('Joint Plot: Motor Current vs. Oil Temperature')
plt.show()
```

```python
# Dist Plot
plt.figure(figsize=(8, 6))
sns.distplot(df['DV_pressure'])
plt.title('Dist Plot: DV Pressure')
plt.show()
```

```python
# Dist Plot
plt.figure(figsize=(8, 6))
sns.distplot(df['TP2'])
plt.title('Dist Plot: TP2')
plt.show()
```

```python
# Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Pressure_switch', data=df)
plt.title('Count Plot: Pressure Switch')
plt.show()
```

```python
# Count Plot
# plt.figure(figsize=(8, 6))
sns.countplot(x='Oil_level', data=df)
plt.title('Count Plot: Oil_level')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Reservoirs', y='Oil_temperature', data=df)
plt.title('Violin Plot: Reservoirs vs. Oil Temperature')
plt.show()
```
